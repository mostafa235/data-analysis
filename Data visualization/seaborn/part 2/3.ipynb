```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("J:/Data science/ML/Pyhon for ML/3.1 UNZIP_ME_FOR_NOTEBOOKS_V4/DATA/StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4),dpi=130)
sns.boxplot(x='race/ethnicity' , y='math score' , data=df);
```

```python
plt.figure(figsize=(11,5),dpi=180)
sns.set(style='whitegrid')
sns.boxplot(x='race/ethnicity',y='math score' , hue='gender' , data=df);
```

```python
plt.figure(figsize=(11,5),dpi=180)
sns.set(style='whitegrid')
sns.boxplot(x='race/ethnicity',y='math score' , hue='gender' , width= .3, data=df);
```

```python
sns.swarmplot(data=df,x='math score')
```

```python
sns.swarmplot(data=df,x='math score',size=2);
```
