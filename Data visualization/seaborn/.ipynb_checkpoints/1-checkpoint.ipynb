```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("J:\student-alcohol-consumption.csv" , index_col=0)
df.head(5)
```

```python
sns.scatterplot(x= "absences" , y = "G3" , data=df , hue ="location")
plt.show()
```

```python
sns.scatterplot(x="absences",y="G3" ,data=df,hue="location" , hue_order=["Rural","Urban"])
plt.show()
```

```python
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x="school" , data=df ,hue="location" ,palette = palette_colors)

# Display plot
plt.show()
```

```python
sns.relplot(x="absences" , y = "G3", data=df, kind="scatter")
plt.show()
```

```python
filt = df["absences"]>70
df[filt]
```

```python
sns.relplot(x="absences" , y = "G3", data=df, kind="scatter" , col="study_time")
plt.show()
```

```python
sns.relplot(x="absences" , y = "G3", data=df, kind="scatter" , row="study_time")
plt.show()
```

```python
sns.relplot(x="G1",y="G3",data=df , kind="scatter")
plt.show()
```

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=df,
            kind="scatter", 
            row="study_time")

# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=df,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row="famsup",
            row_order=["yes", "no"])

# Show plot
plt.show()
```
