```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [5,10,15,20]
y = [55,60,70,85]
```

```python
fig, ax = plt.subplots()
ax.plot(x,y,color = "r")
ax.set_xlabel("(x)")
ax.set_ylabel("(y)")
ax.set_title("title")
plt.show()
```

```python
x1 = [5,10,15,20]
y1 = [55,60,70,85]
x2 = [15,20,30,41]
y2 = [55,60,70,66]
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(x1,y1)
ax[1].plot(x2,y2)
plt.show()
```
