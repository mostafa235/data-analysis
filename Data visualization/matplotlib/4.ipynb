```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
climate_change = pd.read_csv("J:\climate_change.csv" , parse_dates=["date"] , index_col="date")
climate_change.head(3)
```

```python
fig , ax = plt.subplots()
ax.scatter(climate_change["co2"] , climate_change["relative_temp"])
plt.show()
```

```python
fig , ax = plt.subplots()
egiht = climate_change["1980-01-01" : "1989-31-12"]
nine = climate_change["1990-01-01" : "1999-31-12"]
ax.scatter(egiht["co2"] ,egiht["relative_temp"] ,color = "r" , label="egiht")
ax.scatter(nine["co2"] ,nine["relative_temp"] ,color = "b" , label="nine")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

# Add data: "co2", "relative_temp" as x-y, index as color
ax.scatter(climate_change["co2"],climate_change["relative_temp"] ,c = climate_change.index)

# Set the x-axis label to "CO2 (ppm)"
ax.set_xlabel("CO2 (ppm)")

# Set the y-axis label to "Relative temperature (C)"
ax.set_ylabel("Relative temperature (C)")

plt.show()
```
