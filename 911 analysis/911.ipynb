```python
def map_hours(x):
    if x >= 0 and x<12:
        return 'Morning'
    elif x >=12 and x<19:
        return 'AfterNoon'
    else:
        return 'Evening'
    
df['period']=df['hour'].apply(map_hours)
df
```

```python
def map_month(x):
    if x in [12,1,2]:
        return 'Winter'
    elif x in [3,4,5]:
        return 'Spring'
    elif x in [6,7,8]:
        return 'Summer'
    else:
        return 'Autumn'
    
df['season']=df['month'].apply(map_month)
df
```

```python
sns.countplot(x='season',data=df,hue='Reason')
```

```python
sns.catplot(x='season',data=df,hue='Reason',row='Year',kind='count')
```
