```python
df['Time']=pd.to_datetime(df['Time'],format='%m/%d/%Y %H:%M')
df
```

```python
df.info()
```

```python
df['year']=df['Time'].dt.year
df
```

```python
df['month']=df['Time'].dt.month
df['month_name']=df['Time'].dt.month_name()
df['day']=df['Time'].dt.day
df['day_name']=df['Time'].dt.day_name()
df['hour']=df['Time'].dt.hour
df['minute']=df['Time'].dt.minute
df
```

```python
#Seasons 12--- winter,Spring,Summer,Autumn
#periods  [0:12]Morning,[12:19]AfterNoon,[19:24]Evening

def ext_season(x):
    if x in [12,1,2]:
        return 'Winter'
    elif x in [3,4,5]:
        return 'Spring'
    elif x in [6,7,8]:
        return 'summer'
    else:
        return 'Autumn'
    
df['season']=df['month'].apply(ext_season)
df
```

```python
def ext_period(x):
    if x in range(0,12):
        return 'Morning'
    elif x in range(12,19):
        return 'AfterNoon'
    else:
        return 'Evening'

df['Period']=df['hour'].apply(ext_period)
df
```
