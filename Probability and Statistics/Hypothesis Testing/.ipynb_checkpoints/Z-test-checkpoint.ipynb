```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_feather("stack_overflow.feather")
df.head(2)
```

```python
hyp_mean = 110000
```

```python
sample_mean = df['converted_comp'].mean()
sample_mean
```

```python
boot_dist = []
for i in range(5000):
    boot_dist.append(
        np.mean(df.sample(frac=1,replace=True)['converted_comp']))
```

```python
boot_dist
```

```python
import matplotlib.pyplot as plt
plt.hist(boot_dist,bins=50)
plt.show()
```

```python
std_error = np.std(boot_dist,ddof=1)
std_error
```

```python
z_score = (sample_mean-hyp_mean) / std_error
z_score
```

```python
from scipy.stats import norm
p_value = 1 - norm.cdf(z_score , loc=0 , scale = 1)
p_value
```

```python
alpha = 0.05/2
print(p_value <= alpha)
```
